### 1. Which establishments have a hygiene score equal to 20?

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query =  {'LocalAuthorityName':'City of London Corporation','RatingValue': {'$in':['4','5']}}
fields = {'BusinessName':1,'BusinessType':1,'BusinessTypeID':1, 'RatingValue':1}

# Use count_documents to display the number of documents in the result
results = establishments.count_documents(query)
print(f'Number of Documents with London as the Local Authority \nand has a RatingValue greater than or equal to 4: {results}')

# Display the first document in the results using pprint
results_doc = establishments.find_one(query)
pprint(results_doc)

```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

rest_lat = 51.49014200
rest_lon = 0.08384000
degree_search = 0.01
latitude = {'$gte': rest_lat - degree_search , '$lte': rest_lat + degree_search}
longitude = {'$gte': rest_lon - degree_search , '$lte': rest_lon + degree_search}

query = {
        'geocode.longitude': longitude
        ,'geocode.latitude': latitude
        #'geocode' : { '$near' : {'$geometry': { 'type': "Point",  'coordinates': [  0.08384000, 51.49014200] },'$maxDistance': (1000)}}
        
        ,'RatingValue': '5'
       }
fields = {'BusinessName':1,'BusinessType':1,'BusinessTypeID':1, 'RatingValue':1, 'scores.Hygiene':1
          ,'HowClose': {'$abs':{'$sqrt':{'$sum':[{'$pow':[{'$subtract':  ['$geocode.latitude',51.49014200]}, 2]},{'$pow':[{'$subtract':  ['$geocode.longitude',0.08384000]}, 2]}]}}}
          , 'geocode.longitude':1,'geocode.latitude':1}
        
sort =  [('scores.Hygiene',1)]
limit = 5
# Print the results
top_5_results = list(establishments.find(query,fields).sort(sort))

pprint(top_5_results)
```

```python
# Convert result to Pandas DataFrame
results_doc = establishments.find(query,fields)
results_analysis_3=(list(top_5_results))
results_analysis_dff = pd.DataFrame(results_analysis_3[0:5])
results_analysis_dff
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match':
                {
                    'scores.Hygiene': 0
                }
                }
# 2. Groups the matches by Local Authority
group_query = {'$group':{'_id':'$LocalAuthorityName', 'count': {'$sum':1}}}
# 3. Sorts the matches from highest to lowest
sort_values = {'$sort':{'count': -1}}

pipeline = [match_query, group_query, sort_values]
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print('Number of documents:', len(results))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
results = list(establishments.aggregate(pipeline))
results_analysis_4=(list(results))
results_analysis_dff = pd.DataFrame(results_analysis_4)
results_analysis_dff

# Display the number of rows in the DataFrame
print('Rows in DataFrame: ', len(results_analysis_dff))

# Display the first 10 rows of the DataFrame
results_analysis_dff.head(10)
```
